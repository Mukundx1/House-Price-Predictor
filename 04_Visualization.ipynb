```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('CleanedData.csv')
```

```python
df.head()
```

```python
sns.displot(data=df,x='Amount(in rupees)',kind='hist',height=7)
```

```python
sns.catplot(data=df,x='location',y='Amount(in rupees)',kind='box',aspect=1.5,height=7)
plt.xticks(rotation=90)
```

```python
sns.relplot(data=df,y='Area',x='Amount(in rupees)',kind='scatter',height=7)
```

```python
sns.pairplot(df,hue='location',size=2.5)
```
